```python
import pandas as pd
import requests
```

```python
url = 'https://api.covidtracking.com/v1/states/daily.json'
json = requests.get(url).json()
```

```python
df = pd.DataFrame(json)
df
```

```python
who_df = pd.read_csv('global-data.csv')

who_df
```

```python
us_df = who_df[who_df['Country'] == 'United States of America'] 

us_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

covid_heatmap = who_df.pivot(index= 'Country', columns = 'Date_reported', values = 'New_cases')
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(covid_heatmap, cmap='YlOrRd')

plt.title('COVID-19 New Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Country')

plt.show()
```
