```python
import requests
import pprint
import pandas as pd

url = 'https://api.covidtracking.com/v1/states/daily.json'
json = requests.get(url).json()

df= pd.DataFrame(json)
df
```

```python
date = 20200221
filtered_df = df[df['date']==date]
filtered_df
```
