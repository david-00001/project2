```python
from pathlib import Path
Path('covid_data.db').touch()
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('covid_data.db')
c = conn.cursor()
```

```python
#create race data table
c.execute('''CREATE TABLE state_race (Date,State,Cases_Total,Cases_White,Cases_Black,Cases_Latinx,Cases_Asian,Cases_AIAN,Cases_NHPI,Cases_Multiracial,Cases_Other,Cases_Unknown,Cases_Ethnicity_Hispanic,Cases_Ethnicity_NonHispanic,Cases_Ethnicity_Unknown,Deaths_Total,Deaths_White,Deaths_Black,Deaths_Latinx,Deaths_Asian,Deaths_AIAN,Deaths_NHPI,Deaths_Multiracial,Deaths_Other,Deaths_Unknown,Deaths_Ethnicity_Hispanic,Deaths_Ethnicity_NonHispanic,Deaths_Ethnicity_Unknown,White_Cases_per,Black_Cases_per,Latinx_Cases_per,Asian_Cases_per,Native_Cases_per,Pacific_Islander_Cases_per,White_Deaths_per,Black_Deaths_per,Latinx_Deaths_per,Asian_Deaths_per,Native_Deaths_per,Pacific_Islander_Deaths_per)''')
```

```python
import pandas as pd
#upload data to race data table
# load the data into a Pandas DataFrame
users = pd.read_csv('race_data.csv')
# write the data to a sqlite table
users.to_sql('state_race', conn, if_exists='append', index = False)
# make sure all the data is uploaded
c.execute('''SELECT * FROM state_race''').fetchall()
```

```python
#create state_summary table
c.execute('''CREATE TABLE state_summary (Date,State,Cases_Total,Cases_White,Cases_Black,Cases_Latinx,Cases_Asian,Cases_AIAN,Cases_NHPI,Cases_Multiracial,Cases_Other,Cases_Unknown,Cases_Ethnicity_Hispanic,Cases_Ethnicity_NonHispanic,Cases_Ethnicity_Unknown,Deaths_Total,Deaths_White,Deaths_Black,Deaths_Latinx,Deaths_Asian,Deaths_AIAN,Deaths_NHPI,Deaths_Multiracial,Deaths_Other,Deaths_Unknown,Deaths_Ethnicity_Hispanic,Deaths_Ethnicity_NonHispanic,Deaths_Ethnicity_Unknown,White_Cases_per,Black_Cases_per,Latinx_Cases_per,Asian_Cases_per,Native_Cases_per,Pacific_Islander_Cases_per,White_Deaths_per,Black_Deaths_per,Latinx_Deaths_per,Asian_Deaths_per,Native_Deaths_per,Pacific_Islander_Deaths_per,Total_Population,White,White_per,Black,Black_per,Native_American_or_Alaskan_Native,Native_American_or_Alaskan_Native_per,Asian,Asian_per,Pacific_Islander,Pacific_Islander_per,Other,Other_per,Mixed,Mixed_per,Latinx,Latinx_per)''')
```

```python
#upload data to state_summy table

# load the data into a Pandas DataFrame
users = pd.read_csv('summary_stats.csv')
# write the data to a sqlite table
users.to_sql('state_summary', conn, if_exists='append', index = False)
# make sure all the data is uploaded
c.execute('''SELECT * FROM state_summary''').fetchall()
```
