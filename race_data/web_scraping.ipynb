```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Making a GET request
r = requests.get('https://en.m.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_race/ethnicity')
 
# check status code for response received
# success code - 200
print(r)
 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
# testing finding html
table = soup.find_all('tr')
print(table[0].find_all('th')[0].text)
```

```python
# creating a list of the column headers

table_head = soup.find_all('th')
column_head = []
for column in table_head:
    head = column.text
    title = head.strip()
    column_head.append(title)

# Renaming column heads to ensure there are no duplicates
column_head[2]= "White"
column_head[3]= "White_per"
column_head[4]= "Black"
column_head[5]= "Black_per"
column_head[6]= "Native_American_or_Alaskan_Native"
column_head[7]= "Native_American_or_Alaskan_Native_per"
column_head[8]= "Asian"
column_head[9]= "Asian_per"
column_head[10]= "Pacific_Islander"
column_head[11]= "Pacific_Islander_per"
column_head[12]= "Other"
column_head[13]= "Other_per"
column_head[14]= "Mixed"
column_head[15]= "Mixed_per"
column_head[16]= "Latinx"
column_head[17]= "Latinx_per"
column_head
```

```python
# scrapping the table contents by row, adding to row dictionary, and then appending to state list. 
# column_head is used as the column header
table_row = soup.find_all('tr')

states = []
for i in range(1,59):
    row = {}
    row[column_head[0]] = table_row[i].find_all('td')[0].text.strip()
    row[column_head[1]] = table_row[i].find_all('td')[1].text.strip()
    row[column_head[2]] = table_row[i].find_all('td')[2].text.strip()
    row[column_head[3]] = table_row[i].find_all('td')[3].text.strip()
    row[column_head[4]] = table_row[i].find_all('td')[4].text.strip()
    row[column_head[5]] = table_row[i].find_all('td')[5].text.strip()
    row[column_head[6]] = table_row[i].find_all('td')[6].text.strip()
    row[column_head[7]] = table_row[i].find_all('td')[7].text.strip()
    row[column_head[8]] = table_row[i].find_all('td')[8].text.strip()
    row[column_head[9]] = table_row[i].find_all('td')[9].text.strip()
    row[column_head[10]] = table_row[i].find_all('td')[10].text.strip()
    row[column_head[11]] = table_row[i].find_all('td')[11].text.strip()
    row[column_head[12]] = table_row[i].find_all('td')[12].text.strip()
    row[column_head[13]] = table_row[i].find_all('td')[13].text.strip()
    row[column_head[14]] = table_row[i].find_all('td')[14].text.strip()
    row[column_head[15]] = table_row[i].find_all('td')[15].text.strip()
    row[column_head[16]] = table_row[i].find_all('td')[16].text.strip()
    row[column_head[17]] = table_row[i].find_all('td')[17].text.strip()
    states.append(row)
states
```

```python
# creating a dataframe fromt the states list
df = pd.DataFrame(states)
df.dtypes
```

```python
# changing the states name to the abbreviation
states_abb =['AL','AK','AZ','AR','CA','CO','CT','DE','DC','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY','AS','GU','MP','PR','VI']
df.iloc[range(2,58),0] = states_abb
```

```python
# removing percentage sign from percentage columns
df = df.replace("%","",regex=True)
```

```python
# making sure there aren't any null values, then coverting to a float and dividing by 100 to get decimal value
df['White_per'] = df['White_per'].replace('','0').astype('float')/100
df['Black_per'] = df['Black_per'].replace('','0').astype('float')/100
df['Native_American_or_Alaskan_Native_per'] = df['Native_American_or_Alaskan_Native_per'].replace('','0').astype('float')/100
df['Asian_per'] = df['Asian_per'].replace('','0').astype('float')/100
df['Pacific_Islander_per'] = df['Pacific_Islander_per'].replace('','0').astype('float')/100
df['Other_per'] = df['Other_per'].replace('','0').astype('float')/100
df['Mixed_per'] = df['Mixed_per'].replace('','0').astype('float')/100
df['Latinx_per'] = df['Latinx_per'].replace('','0').astype('float')/100
df.rename(columns = {'Total Population':'Total_Population'},inplace=True)

```
