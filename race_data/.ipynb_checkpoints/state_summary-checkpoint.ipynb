```python
import pandas as pd
```

```python
# defining csv paths
race_csv = 'race_data.csv'
state_csv = 'state_pop.csv'

# creating dataframes from csv's
race_df = pd.read_csv(race_csv)
state_df = pd.read_csv(state_csv)
```

```python
# checking the number of unique states
len(race_df['State'].unique())
```

```python
# getting the last date of recorded data from the data set
total_df = race_df.loc[race_df['Date']=='2021-03-07']
```

```python
# confirming that there are 56 rows to match the unique states
total_df
```

```python
# merging the two data frames
state_stats = total_df.merge(state_df, how = 'left')
```
