```python
csv = "race_data.csv"
```

```python
df = pd.read_csv(csv)
df = df.sort_values(by=['State','Date'])
```

```python
df.columns
```

```python
df = df[['Date', 'State', 'Cases_Total', 'Cases_White', 'Cases_Black',
       'Cases_Latinx', 'Cases_Asian', 'Cases_AIAN', 'Cases_NHPI',
       'Cases_Multiracial', 'Cases_Other', 'Cases_Unknown',
       'Cases_Ethnicity_Hispanic', 'Cases_Ethnicity_NonHispanic',
       'Cases_Ethnicity_Unknown', 'Deaths_Total', 'Deaths_White',
       'Deaths_Black', 'Deaths_Latinx', 'Deaths_Asian', 'Deaths_AIAN',
       'Deaths_NHPI', 'Deaths_Multiracial', 'Deaths_Other', 'Deaths_Unknown',
       'Deaths_Ethnicity_Hispanic', 'Deaths_Ethnicity_NonHispanic',
       'Deaths_Ethnicity_Unknown']]
```

```python
# replace none and Nan
df
```

```python
df['White_Cases_per']= df['Cases_White']/df['Cases_Total']
df['Black_Cases_per']= df['Cases_Black']/df['Cases_Total']
df['Latinx_Cases_per']= df['Cases_Latinx']/df['Cases_Total']
df['Asian_Cases_per']= df['Cases_Asian']/df['Cases_Total']
df['Native Cases_per']= df['Cases_AIAN']/df['Cases_Total']
df['Pacific Islander Cases_per']= df['Cases_NHPI']/df['Cases_Total']

df['White Deaths_per']= df['Deaths_White']/df['Deaths_Total']
df['Black Deaths_per']= df['Deaths_Black']/df['Deaths_Total']
df['Latinx Deaths_per']= df['Deaths_Latinx']/df['Deaths_Total']
df['Asian Deaths_per']= df['Deaths_Asian']/df['Deaths_Total']
df['Native Deaths_per']= df['Deaths_AIAN']/df['Deaths_Total']
df['Pacific Islander Deaths_per']= df['Deaths_NHPI']/df['Deaths_Total']
df.dtypes
```

```python
df.to_csv('race_data.csv', index=False)
```
